**Tratamiento NA: Variables numéricas**

Los tratamientos para la sustitución de valores NA, pueden enfocarse en varias estrategias: 

- *Eliminar valores NA:* Esta opción en nuestro caso queda descartada ya que hay más de 200 filas con valores NA. Lo que supondría perder casi el 50% de nuestras observaciones.

- *Sustituir por la media o mediana:* En el caso de las variables continuas, podemos sustituir por el valor de la media o mediana (dependiendo de la distribución de la variable en si).

- *Sustituir por el valor común:* Para las variables no continuas, se puede optar por la sustitución de los valores NA por el valor de la clase más prominente.

Como vemos, en nuestro caso optaremos por la segunda opción (media) para la variable 'wt_before_delivery' y la tercera opción (clase más prominente) para las otras dos variables numéricas 'nº of sibling' y 'miscarrage history'. 

```python
# Tratamiento NA para variable wt_before_delivery en X_train:
meanwt = X_train['wt_before_delivery'].mean()
X_train['wt_before_delivery'].fillna(meanwt, inplace=True)

# Comprobamos si ha ido bien y hemos eliminado los valores NA: 
X_train['wt_before_delivery'].isna().sum()
```

```python
# Tratamiento NA para variable wt_before_delivery en X_test:
meanwt2 = X_test['wt_before_delivery'].mean()
X_test['wt_before_delivery'].fillna(meanwt2, inplace=True)

# Comprobamos si ha ido bien y hemos eliminado los valores NA: 
X_test['wt_before_delivery'].isna().sum()
```

```python
# Tratamiento NA para para la variable No_of_sibling en X_train:
X_train['No_of_sibling']= X_train['No_of_sibling'].fillna(X_train['No_of_sibling'].mode().iloc[0])

# Comprobamos si ha ido bien y hemos eliminado los valores NA: 
X_train['No_of_sibling'].isna().sum() 
```

```python
# Tratamiento NA para para la variable No_of_sibling en X_test:
X_test['No_of_sibling']= X_test['No_of_sibling'].fillna(X_test['No_of_sibling'].mode().iloc[0])

# Comprobamos si ha ido bien y hemos eliminado los valores NA: 
X_test['No_of_sibling'].isna().sum() 
```

```python
# Tratamiento NA para para la variable Miscarrage History en X_train:
X_train['Miscarrage History']= X_train['Miscarrage History'].fillna(X_train['Miscarrage History'].mode().iloc[0])

# Comprobamos si ha ido bien y hemos eliminado los valores NA: 
X_train['Miscarrage History'].isna().sum() 
```

*Paso 3: Determinación grupos normal y outlier*

Repasamos las características de la estadística descriptiva de cada grupo (normal - outlier). Así, podremos ir ajustando nuestro modelo conforme si las medias coinciden con los datos previos. 

```python
# Asignamos a una variable, el valor del threshold obtenido: 
threshold1 = pca1.threshold_
threshold1

# Definimos una función que nos calcule la cantidad de valores outlier:
def descriptive_stat_threshold(df,pred_score, threshold):
    # Definimos un nuevo dataframe:
    df = pd.DataFrame(df)
    # Asignamos a la columna 'Anomaly_Score', el valor de los outliers:
    df['Anomaly_Score'] = pred_score
    # Asignamos a la variable grupo que los valores de anomalía menores al threshold son outliers:
    df['Group'] = np.where(df['Anomaly_Score']< threshold1, 'Normal', 'Outlier')

    # Añadimos al dataframe distintas variables:
    cnt = df.groupby('Group')['Anomaly_Score'].count().reset_index().rename(columns={'Anomaly_Score':'Count'})
    cnt['Count %'] = (cnt['Count'] / cnt['Count'].sum()) * 100 
    stat = df.groupby('Group').mean().round(2).reset_index() 
    stat = cnt.merge(stat, left_on='Group',right_on='Group') 
    return (stat)

descriptive_stat_threshold(X_train1o,y_train_scores1, threshold1)
```

Resultados que nos ofrece la tabla:

- *Tamaño de los grupos:* Podemos tener una indicación de la cantidad de outliers que encontramos en nuestro conjunto de datos. Nos sirve para saber si el threshold ha sido el correcto o tenemos que ir modificando este valor o el "contamination_ratio". En nuestro caso obtenemos que 36 de las 360 observaciones de nuestro conjunto de datos son outliers, un 10% de la muestra.
- *Medias de cada grupo:* Podemos comprobar si las medias del grupo normal o grupo anómalo coinciden o son diferentes (conforme lo esperamos).Aquí podemos ver que las medias de los outliers son diferentes con respecto a los grupos normales, indicando que su distribución difiere con la distribución normal de las muestras.
- *Anomaly_score:* Este valor debe ser mayor en el grupo de outlier que en el normal, lo que se confirma también en nuestro caso. 

Para poder estudiar con más detalle los resultados de este modelo, podemos calcular una matriz de confusión usando los valores de y_test. 

```python
# Matriz confusión _ X_train1:
Actual_pred = pd.DataFrame({'Actual': y_train['PreTerm'], 'Anomaly_Score': y_train_scores1})
Actual_pred['Pred'] = np.where(Actual_pred['Anomaly_Score']< threshold1,0,1)
pd.crosstab(Actual_pred['Actual'],Actual_pred['Pred'])
```

*Interpretación de matriz:* Aquí podemos observar que solamente 5 de los 36 valores reconocidos anteriormente como outliers son identificados como auténticos outliers. Además, sólo 26 4son identificados como valores normales (de 324). Tenemos 31 falsos outliers que se reconocen como outliers pero que realmente son valores normales y 60 falsos *inliers* que se reconocen como outliers. 

Podemos concluir que este método no es del todo preciso, ya que la cantidad de *True Positive y True Negative* es bastante inferior a lo esperado y no concuerda con los valores reales.

```python
# Asignamos a una variable, el valor del threshold obtenido: 
thresholdI3 = isftI3.threshold_

# Definimos una función que nos calcule la cantidad de valores outlier:
def descriptive_stat_threshold(df,pred_score, threshold):
    # Definimos un nuevo dataframe:
    df = pd.DataFrame(df)
    # Asignamos a la columna 'Anomaly_Score', el valor de los outliers:
    df['Anomaly_Score'] = pred_score
    # Asignamos a la variable grupo que los valores de anomalía menores al threshold son outliers:
    df['Group'] = np.where(df['Anomaly_Score']< thresholdI3, 'Normal', 'Outlier')

    # Añadimos al dataframe distintas variables:
    cnt = df.groupby('Group')['Anomaly_Score'].count().reset_index().rename(columns={'Anomaly_Score':'Count'})
    cnt['Count %'] = (cnt['Count'] / cnt['Count'].sum()) * 100 
    stat = df.groupby('Group').mean().round(2).reset_index() 
    stat = cnt.merge(stat, left_on='Group',right_on='Group') 
    return (stat)

descriptive_stat_threshold(X_train3,y_train_scoresI3, thresholdI3)
```

Resultados que nos ofrece la tabla:

- *Tamaño de los grupos:* Podemos tener una indicación de la cantidad de outliers que encontramos en nuestro conjunto de datos. Nos sirve para saber si el threshold ha sido el correcto o tenemos que ir modificando este valor o el "contamination_ratio".
- *Medias de cada grupo:* Podemos comprobar si las medias del grupo normal o grupo anómalo coinciden o son diferentes (conforme lo esperamos).
- *Anomaly_score:* Este valor debe ser mayor en el grupo de outlier que en el normal. 

Para poder estudiar con más detalle los resultados de este modelo, podemos calcular una matriz de confusión usando los valores de y_test.

```python
# Matriz de confusión:
def confusion_matrix(actual,score, threshold):
    Actual_pred = pd.DataFrame({'Actual': actual, 'Pred': score})
    Actual_pred['Pred'] = np.where(Actual_pred['Pred']<=thresholdI3,0,1)
    cm = pd.crosstab(Actual_pred['Actual'],Actual_pred['Pred'])
    return (cm)
confusion_matrix(y_train['PreTerm'],y_train_scoresI3,thresholdI3)
```

*Interpretación de matriz:* Aquí podemos observar que solamente 6 de los 36 valores reconocidos anteriormente como outliers son identificados como auténticos outliers. Además, sólo 265 son identificados como valores normales (de 324). Tenemos 30 falsos outliers que se reconocen como outliers pero que realmente son valores normales y 59 falsos *inliers* que se reconocen como outliers. 

Podemos concluir que este método no es del todo preciso, ya que la cantidad de *True Positive y True Negative* es bastante inferior a lo esperado y no concuerda con los valores reales.

**BOSQUES ALEATORIOS:**

Bosques aleatorios es un algoritmo de aprendizaje automático supervisado que se usa ampliamente en problemas de clasificación y regresión. Construye árboles de decisión en diferentes muestras y toma su voto mayoritario para la clasificación y el promedio en caso de regresión.  

Un modelo Random Forest está formado por un conjunto (ensemble) de árboles de decisión individuales, cada uno entrenado con una muestra aleatoria extraída de los datos de entrenamiento originales mediante bootstrapping). Esto implica que cada árbol se entrena con unos datos ligeramente distintos. En cada árbol individual, las observaciones se van distribuyendo por bifurcaciones (nodos) generando la estructura del árbol hasta alcanzar un nodo terminal. La predicción de una nueva observación se obtiene agregando las predicciones de todos los árboles individuales que forman el modelo.


Las características principales [8,9]:

- *N_estimators:* Número de árboles que el algoritmo construye antes del promedio de las predicciones. 
- *Max_features:* Número máximo de variables del modelo que se consideran para dividir un nodo.
- *Mini_sample_leaf:* Determina el número mínimo de hojas que se requieren para dividir un nodo interno. 
- *Criterion:* Criterio en el que se basa la separación de un nodo, tenemos teoría Gini, entropía o *Log loss*. 
- *Max_leaf_nodes:* Máximo de nodos de hoja en cada árbol.

En nuestro caso, vamos a dejar los valores preestablecidos del algoritmo y observar si así la precisión del modelo llega a nuestros niveles preestablecidos. Sino, modificaremos los hiperparámetros antes mencionados.

**MÁQUINAS DE VECTORES DE SOPORTE:**
    
Es uno de los algoritmos más utilizados dentro del mundo del aprendizaje automático. Es un algoritmo poderoso, fácil de explicar y se generaliza bien en muchos casos (tantos datos categóricos como numéricos). Primero mapea cada elemento de datos en un espacio de características n-dimensional donde n es el número de características. Luego identifica el hiperplano que separa los elementos de datos en dos clases mientras maximiza la distancia marginal para ambas clases y minimiza los errores de clasificación. 


Las características principales para la definición de este modelo [10,11]:

- Kernel: La función principal del kernel es transofrmar los datos de entrada del conjunto de datos dados en la forma requerida. Existen diferentes tipos de funciones como lineal, polinomial y radial. En nuestro caso, empezaremos por un kernel lineal. 
- Regularización: Corresponde al parámetro C de las máquinas de vectores de soporte. El término de clasificacion errónea le dice a la optimización del algoritmo cuánto error es soportable. Un valor pequeño crea un hiperplano de margen pequeño y un valor mayor de C crea un hiperplano de margen mayor. 
- Gamma: Un valor más bajo de Gamma se ajustará libremente al conjunto de datos de entrenamiento, mientras que un valor más alto de gamma se ajustará exactamente al conjunto de datos de entrenamiento, lo que provoca un ajuste excesivo.

**REDES NEURONALES:**

Este algoritmo de aprendizaje automático se llama así ya que se replica la estructura de un cerebro, presentando un grupo de nodos interconectados. La salida de uno de los nodos va como entrada a otro nodo para su posterior procesamiento según la interconexión. Los nodos normalmente se agrupan en una matriz llamada capa dependiendo de la transformación que realicen. Además de la capa de entrada y salida, puede haber una o más capas ocultas. 

Es una de las herramientas más potentes de aprendizaje profundo ya que se pueden detectar relaciones no lineales complejas entre las variables dependientes e independientes, permite adaptarse a problemas de clasificación y regresión. Las características de las redes neuronales que vamos a tener en cuenta para nuestro algoritmo son [12]: 

- Capas definidas: Aunque existan modelos de una sola capa, se ha visto que la combinación de distintas capas ocultas nos permiten mejorar las predicciones del modelo. Es por ello, que estableceremos una capa de entrada que coincida con las variables de cada conjunto de datos, 3 capas ocultas y una capa de salida.

- Función de activación de cada capa: Estas funciones controlan en gran medida que información se propaga desde una capa a la siguiente. En nuestro caso, vamos a usar la función de activación ReLU que aplica una transformación no lineal muy simple.

- Función de coste: Es la encargada de cuantificar la distancia entre el valor real y el valor predicho por la red. Se debe diferenciar entre problemas de regresión o problemas de clasificación para elegir la función correcta. En nuestro caso, al ser un problema de clasificación y tener como valores de 'target' (0,1) elegiremos "binary_crossentropy".

En este caso, convertimos nuestros conjuntos de datos de DataFrame a array tipo numpy ya que la transformación no viene automática en el algoritmo. 

**MODELO ANN: KERAS**

```python
# Importamos las librerías necesarias para la definición e implementación del modelo:
import tensorflow as tf
import tempfile
from tensorflow import keras

from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import InputLayer
from tensorflow.keras.layers import Dropout
```

Se procede a definir el modelo de la red neuronal para df_scaled:

```python
# Definición del modelo y las medidas de precisión:
# Se fija semilla
tf.random.set_seed(123)

# Se definen las medidas
METRICS = [
      keras.metrics.Precision(name='precision'),
      keras.metrics.Recall(name='recall'),
      keras.metrics.AUC(name='auc')
]

# Se crea una función para la definición del modelo de red neuronal:
def make_model(metrics=METRICS, output_bias=None):
  if output_bias is not None:
    output_bias = tf.keras.initializers.Constant(output_bias)
  model = Sequential()
  model.add(InputLayer(input_shape = (df_scaled.shape[1])))
  model.add(Dense(10, activation = 'relu'))     
  model.add(Dense(12, activation = 'relu'))
  model.add(Dense(20, activation = 'relu'))
  model.add(Dropout(0.5))
  model.add(Dense(1, activation = 'sigmoid', bias_initializer=output_bias))

  model.compile(
      optimizer=keras.optimizers.Adam(learning_rate=1e-3),
      loss=keras.losses.BinaryCrossentropy(),
      metrics=METRICS)

  return model
```

*Construcción del modelo:*

Se usa un valor de batch de 2048 para garantizar que cada lote tenga una probabilidad decente de contener algunas muestras positivas. Si el tamaño del lote fuera demasiado pequeño, probablemente no se tendría los valores de partos prematuros para poder aprender.

Además, añadimos la opción de *early stopping* para cuando el modelo llegue a un valor constante de f1 o pérdida de error, se pare de calcular el modelo. Evitando así problemas como *overfitting*.

```python
# Se definen los valores básicos del modelo (epoch y batch):
EPOCHS = 100
BATCH_SIZE = 2048

# Definimos una función de "early stopping":
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_prc', 
    verbose=1,
    patience=10,
    mode='max',
    restore_best_weights=True)
```

```python
# Mostramos el modelo definido para df_scaled1 y sus características:
model = make_model()
model.summary()
```
